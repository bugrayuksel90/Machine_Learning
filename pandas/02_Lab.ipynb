```python
# UnitPrice 18'den büyük olan verilerin ProductID, ProductName, UnitPrice, UnitsInStock, sütunlarını listeleyin
#path I daha rahat okunur
(df.query('UnitPrice >18.00')
    .loc[:, ['ProductID', 'ProductName',  'UnitPrice', 'UnitsInStock']]
    .sort_values(by='UnitPrice', ascending=False))

#path II minik bir performans farkı sunar
# df[
#     df['UnitPrice'] > 18.00
# ][['ProductID', 'ProductName',  'UnitPrice', 'UnitsInStock']].sort_values(by='UnitPrice', ascending=False)

#path III
# df[
#     df['UnitPrice'] > 18.00
# ].loc[:, ['ProductID', 'ProductName',  'UnitPrice', 'UnitsInStock']].sort_values(by='UnitPrice', ascending=False)
```

```python
# UnitPrice 20'den büyük ve UnitsInStock'u 100 den büyük olan verilerin ProductID, ProductName, UnitPrice, UnitsInStock, sütunlarını listeleyin

(df.query('UnitPrice > 20.00 and UnitsInStock > 100')
    .loc[:, ['ProductID', 'ProductName',  'UnitPrice', 'UnitsInStock']]
    .sort_values(by='UnitPrice', ascending=False))
```

```python
# QuantityPerUnit sütunun da boxes ifadesi geçen ürünlerin ProductID, ProductName, UnitPrice, UnitsInStock, sütunlarını listeleyin

(df.query('QuantityPerUnit.str.contains("boxes")')
    .loc[:, ['ProductID', 'ProductName','QuantityPerUnit', 'UnitPrice', 'UnitsInStock']]
    .sort_values(by='UnitPrice', ascending=False))
```

```python
# QuantityPerUnit sütunun da boxes ifadesi geçen  ve CategoryID si 2 olan ürünlerin ProductID, CategoryID, ProductName, UnitPrice, UnitsInStock,
#  sütunlarını listeleyin

(df.query('QuantityPerUnit.str.contains("boxes") and CategoryID == 2')
    .loc[:, ['ProductID', 'CategoryID', 'ProductName','QuantityPerUnit', 'UnitPrice', 'UnitsInStock']]
    .sort_values(by='UnitPrice', ascending=False))
```

```python
df_orders = pd.read_csv(filepath_or_buffer='data/orders.csv')
df_orders[:5]
```

```python
# sipariş aralığın 01/07/1996 ile  31/07/1996 aralığındaki olan siprişlerin OrderID, CustomerID, OrderDate, ShipName listeleyin
(
    df_orders.query('OrderDate >= "01/07/1996" and OrderDate <= "31/07/1996"')
        .loc[:, ['OrderID', 'CustomerID', 'OrderDate','ShipName']]
        .sort_values(by='CustomerID', ascending=True)
)
```
